```python
Fraud = data[data['Class'] == 1]
Normal = data[data['Class'] == 0]
Fraud.head()
```

```python
#画多图，多图变量名为f, 第一个图变量名ax1,第二个变量名ax2,plt.subplot函数，参数为xyz'或者x, y, z.sharex共享X坐标轴
f, (ax1, ax2) = plt.subplots(2, 1, sharex = False)
plt.figure(figsize=(12,10))
f.suptitle('Time transacation va Amount')
ax1.scatter(Fraud.Time, Fraud.Amount)
ax1.set_title('Fraud')
ax2.scatter(Normal.Time, Normal.Amount)
ax2.set_title('Normal')
```

```python
#设置图形尺寸
plt.figure(figsize=(10,8))
#画一个一行两列的图，现在画第一行第一列的图
plt.subplot(121)
Fraud.Amount.plot.hist(title='Fraud')
#现在画第二个图
plt.subplot(122)
Normal.Amount.plot.hist(title='Normal')
```

# 透视表
data.pivot_table(values='Amount', index='hour', columns='Class', aggfunc='count')

```python
import seaborn as sns
```

```python
plt.figure(figsize=(16,8))
sns.barplot('hour', 'Class', data = data)
plt.title('HOURLY Fraud Ratio')
plt.ylabel('Fraud Ratio')
```

```python
plt.figure(figsize=(15,6))
sns.distplot(data[data['Class']==1]['hour'],
            norm_hist=True,
            bins=24,
            kde=True,
            color='b',
            hist_kws={'alpha':0.5})
sns.distplot(data[data['Class']==0]['hour'],
            norm_hist=True,
            bins=24,
            kde=True,
            color='r',
            hist_kws={'alpha':0.5})
```

```python
plt.figure(figsize=(15,6))
sns.distplot(data[data['Class']==1]['hour'],
            norm_hist=True,
            bins=24,
            kde=False,
            color='b',
            hist_kws={'alpha':0.5})
sns.distplot(data[data['Class']==0]['hour'],
            norm_hist=True,
            bins=24,
            kde=False,
            color='r',
            hist_kws={'alpha':0.5})
```
